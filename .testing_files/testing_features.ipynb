```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import datetime
```

```python
df1 = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
```

```python
df1.head()
```

```python
df1["date"] = pd.to_datetime(df1.date).dt.date
df1['date'] = pd.DatetimeIndex(df1.date)
```

```python
df1['7day_rolling_avg_cases'] = df1['new_cases'].rolling(window=7).mean()
```

```python
sns.lineplot(x = 'date',
             y = '7day_rolling_avg_cases', 
             data = df1,
             label = 'Rolling_average')
```

```python
df1['7day_rolling_avg_deaths'] = df1['new_deaths'].rolling(window=7).mean()
```

```python
sns.lineplot(x = 'date',
             y = '7day_rolling_avg_deaths', 
             data = df1,
             label = 'Rolling_average')
```

```python
df1['cumulative_number_cases'] = df1['new_cases'].cumsum(axis = 0)
```

```python
sns.lineplot(x = 'date',
             y = 'cumulative_number_cases', 
             data = df1,
             label = 'Cumulative_number')
```

```python
df1['cumulative_number_deaths'] = df1['new_deaths'].cumsum(axis = 0)
```

```python
sns.lineplot(x = 'date',
             y = 'cumulative_number_deaths', 
             data = df1,
             label = 'Cumulative_number')
```
